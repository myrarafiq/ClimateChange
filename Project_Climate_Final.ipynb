Every year, countries around the world gather at the UN's <a href="https://www.un.org/en/climatechange/cop28">COP meetings</a> to discuss how they, collectively, can reduce carbon emissions. Most countries set an emissions target. However, these targets are widely debated: Should countries that manufacture most of the world's goods be allowed to have higher emissions targets, since their manufacturing is crucial to multiple economies worldwide? Should countries with more urban density be expected to have lower emissions targets, since their citizens tend to travel less? Should countries with extremely cold or warm weather be allowed more emissions for cooling or heating their homes?

In this project, we investigate the factors that are most strongly predictive of a country's emissions. Then, we use our models to predict which countries will be the highest and lowest polluters. In doing so, our models can provide a more informed view of the amount of emissions that can be expected by each country in future years - along with recommended ways to reduce those emissions.

#### The Dataset - `co2_dataset`

**Description**
This dataset is a comprehensive collection of socio-economic and environmental indicators for various countries over multiple years. 

**Variables**
- `country`: name of country 
- `year`: year the data was recorded
- `area`: countries total area in square kilometers
- `agriculture_land_per`: percentage of the total land area used for agriculture
- `population`: total population of the country
- `gdp`: Gross Domestic Product (GDP) measured in current US dollars. GDP is a measure for the overall size of a country's economy.
- `agriculture_gdp_per`: percentage of GDP contributed by the agriculture, forestry, and fishing sector.
- `industry_gdp_per`: percentage of GDP contributed by the industrial sector, including construction
- `manufacturing_gdp_per`: percentage of GDP contributed by the manufacturing sector
- `service_gdp_per`: percentage of GDP contributed by the services sector 
- `import_goods_currency`: total value of goods and services imported, measured in current US dollars
- `import_goods_gdp_per`: percentage of GDP represented by imports
- `export_goods_currency`: total value of goods and services exported, measured in current US dollars
- `export_goods_gdp_per`: percentage of GDP represented by exports
- `trade_gdp_per`: percentage of GDP represented by the total trade (imports plus exports)
- `agriculture_employment_per`:percentage of people working in agriculture related jobs from a countries total employment 
- `industry_employment_per`: percentage of people working in industry related jobs from a countries total employment 
- `service_employment_per`: percentage of people working in service related jobs from a countries total employment 
- `life_expectancy`: average number of years a newborn is expected to live
- `urban_pop_per`: urban population percentage 
- `urban_growth_per`: urban population percentage growth 
- `rural_pop_per`: rural population percentage 
- `rural_growth_per`: rural population percentage growth 
- `co2_emission`: carbon dioxide emission in metric tons **per capita** (per person)

## Part 2: Data Processing & Baseline Model

We will now prepare our data for modeling and fit a baseline model. Our baseline model's accuracy will be used as a comparison point for more advanced models. In this section, we will: 
- Split our data into train and test sets (code provided).
- Fit a baseline model to predict `co2_emission`. This model will simply be the mean of `co2_emission` in the train data (i.e. the mean model).
- Test the prediction accuracy of our baseline model by calculating its RMSE on the test set.

Since we don't have to one-hot encode any of our data, we can go straight to splitting it into a train set and a test set. We'll do an 80-20 train/test split:

```r
# Set seed so that we get the same random sample each time
set.seed(321)

# Get total numder of rows (N) in dataset
N = nrow(co2_data)

# Get szie of train set so that it's 80% of dataset, round to the nearest whole number
train_size = round(0.8 * N)

# Sample random row indeeces from the full dataset, without replacement
sample <- sample.int(n=N, size = train_size, replace = F)

# Select the sampled row indeces for train set, remaining go to test set
train <- co2_data[sample,]
test  <- co2_data[-sample, ]

# Print the dimensions of our train and test sets
dim(train)
dim(test)
```

```r
mean_co2 <- mean(train$co2_emission)
mean_co2
```

```r
error <- test$co2_emission - mean_co2
# error

sqrt(mean(error^2)) #rmse is larger than the mean, not a good model
```
